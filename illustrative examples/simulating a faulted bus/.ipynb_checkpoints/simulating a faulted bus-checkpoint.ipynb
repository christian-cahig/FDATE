## <a id="Imports-and-utilities"></a>   Imports and utilities

```python
import matplotlib, numpy, os, pickle
from matplotlib import cm, pyplot

# Import custom utility module
import sys
sys.path.insert(1, "../utility")
from utils import SimulationData, build_banded_matrix

%matplotlib inline
```

```python
# Experiment parameters
t_fault = 0.3                            # time at which the fault occurs
u_fault = 0.                             # fault voltage
t_cleared_receiving_end = t_fault + 0.4  # time at which the line switch in the receiving end is opened
t_cleared_sending_end = t_fault + 0.5    # time at which the line switch in the sending end is opened
save_dir = "./results"                   # directory to which results are saved
```

## <a id="Preliminaries"></a>   Preliminaries

```python
# Initialize data
data = SimulationData(
    R=1.5e-2, L=1e-3, G=1e-10, C=1e-8,
    X=150e3, T=1.5,
    name="experiment_04"
)
```

```python
# Initial conditions
mu_0 = lambda x : numpy.sin(120*numpy.pi*x)
xi_0 = 0.

data.initial_conditions(mu_0=mu_0, xi_0=xi_0)
print("Initial conditions set.")

# Boundary conditions
def nu_0(t, t_clear, u_1_n, delta_x, gamma_0=0):
    if t < t_clear:
        return 1.5 * numpy.sin(120*numpy.pi*t)
    else:
        return u_1_n - gamma_0 * delta_x

def nu_X(t, t_fault, t_clear, u_fault, u_Kless1_n, delta_x, gamma_X=0):
    if t < t_fault:
        return u_Kless1_n + gamma_X * delta_x
    elif t_fault <= t <= t_clear:
        return u_fault
    else:
        return u_Kless1_n + gamma_X * delta_x

data.boundary_conditions(nu_0=nu_0, nu_X=nu_X)
print("Boundary conditions set.")

# Workspace cleanup
del mu_0, xi_0, nu_0, nu_X
```

```python
# Spatial domain discretization
data.set_domain(num_x_points = 500)
x_domain = data.get_x_domain()

# Temporal domain discretization
delta_t_max = data.domain['x step'] / data.line['c']
delta_t = 0.10*delta_t_max
num_t_points = (data.domain['T'] / delta_t) + 1
num_t_points = int(numpy.round(num_t_points / 1000, 1)) * 1000
data.set_domain(num_t_points = num_t_points)
t_domain = data.get_t_domain()

# Workspace cleanup
del delta_t_max, delta_t, num_t_points
```

```python
# Sanity check
print(f"CFL number = {data.get_CFL_number()}")
```

## <a id="Simulation"></a>   Simulation

```python
# Update scheme parameters
A = 1. + 0.5 * data.domain["t step"] * (data.line["alpha"] + data.line["beta"])
B = 1. - 0.5 * data.domain["t step"] * (data.line["alpha"] + data.line["beta"])
E = data.line["c squared"] * data.domain["t step"]**2 / data.domain["x step"]**2
F = 2. - (2. * data.line["c squared"] * data.domain["t step"]**2 / data.domain["x step"]**2) - (data.line["alpha"] * data.line["beta"] * data.domain["t step"]**2)
E_matrix = build_banded_matrix(numpy.array([E, F, E]), data.domain["K"]-1)
B_matrix = B*numpy.concatenate(
    (numpy.zeros((data.domain["K"]-1,1)), numpy.identity(data.domain["K"]-1), numpy.zeros((data.domain["K"]-1,1))),
    axis=1
)
```

```python
# Intitialize container for voltages
u = numpy.zeros((data.domain["K"]+1, data.domain["N"]+1))
```

```python
# Initial conditions
u[:,0] = data.init_condition_value(x_domain)
u[0,0] = data.dirichlet_sending_end(t_domain[0], t_cleared_sending_end, u[1,0], data.domain['x step'])
u[-1,0] = data.dirichlet_receiving_end(t_domain[0], t_fault, t_cleared_receiving_end, u_fault, u[-2,0], data.domain['x step'])

u[:,1] = data.init_condition_deriv(u[:,0])
u[0,1] = data.dirichlet_sending_end(t_domain[1], t_cleared_sending_end, u[1,1], data.domain['x step'])
u[-1,1] = data.dirichlet_receiving_end(t_domain[0], t_fault, t_cleared_receiving_end, u_fault, u[-2,1], data.domain['x step'])
```

```python
# n = 1,2,...,N-1
for n in range(1,data.domain["N"]):
    u_was, u_now = u[:,n-1], u[:,n]
    
    # k=1,2,...,K-1
    tilde_u_nxt = numpy.dot(E_matrix, u_now) - numpy.dot(B_matrix, u_was)
    tilde_u_nxt /= A
    
    # k=0,1,...,K
    u_nxt = numpy.concatenate(
        (numpy.array([data.dirichlet_sending_end(t_domain[n+1], t_cleared_sending_end, tilde_u_nxt[0], data.domain['x step'])]),
         tilde_u_nxt,
         numpy.array([data.dirichlet_receiving_end(t_domain[n+1], t_fault, t_cleared_receiving_end, u_fault, tilde_u_nxt[-1], data.domain['x step'])]))
    )
    u[:,n+1] = u_nxt

# Workspace cleanup
del n, u_was, u_now, tilde_u_nxt, u_nxt
```

```python
# Save results
result = {
    "CFL number" : data.get_CFL_number(),
    "fault info" : {
        "event time" : t_fault, "sending end cleared" : t_cleared_sending_end,
        "receiving end cleared" : t_cleared_receiving_end, "voltage" : u_fault
    },
    "numerical solution" : u
}
with open(os.path.join(save_dir, "faulted_bus.pkl"), 'wb') as file:
    pickle.dump(result, file)

print("Results saved to " + os.path.join(save_dir, "faulted_bus.pkl."))

# Workspace cleanup
del A, B, E, F, E_matrix, B_matrix
del u, result, file
```

### <a id="Results-summary--Voltage-profile-at-pre-fault-condition"></a>   Voltage profile at pre-fault condition

```python
print("Investigating ", end="")
min_n_to_observe = 0
max_n_to_observe = numpy.argmin(numpy.abs(t_domain - t_fault))
ns_to_observe = numpy.linspace(min_n_to_observe, max_n_to_observe, 8).astype(int)
print(f"at n = {list(ns_to_observe)}")

# Workspace cleanup
del min_n_to_observe, max_n_to_observe
```

```python
# 2D visualization
from matplotlib.ticker import ScalarFormatter
matplotlib.style.use("seaborn")
figure = pyplot.figure(dpi=130)

for i, n_to_observe in enumerate(ns_to_observe):
    axes = figure.add_subplot(4,2,i+1)
    axes.set_title(f"At $t_n = {t_domain[n_to_observe]:.3g}$", fontsize=8.5)
    axes.set_xlabel("$x_k$ (in km)", fontsize=8.5)
    axes.set_ylabel("$u(x_k,t_n)$", fontsize=8.5)
    
    axes.tick_params(labelsize=7)
    axes.yaxis.offsetText.set_fontsize(7)
    axes.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    
    # Plot numerical solutions
    axes.plot(x_domain/1e3, result["numerical solution"][:,n_to_observe], linewidth=0.7)
    
    axes.set_xlim(left=x_domain[0]/1e3, right=x_domain[-1]/1e3)
    axes.grid(True)

figure.tight_layout()
del ns_to_observe, i, n_to_observe, axes
```

```python
# Save summary
figure.savefig(os.path.join(save_dir, "Voltage profiles at pre-fault.png"), dpi=300, facecolor=None, edgecolor=None, bbox_inches="tight", pad_inches=0.05)
```
